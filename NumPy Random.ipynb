```python
import numpy as np
```

```python
#Generating Random numbers

import random
```

```python

#int functions

x= random.randint(10,20)
print(x)

print(random.randrange(10))

#with stepsize=2
print(random.randrange(5,10,2))
```

```python
#float numbers
#since we didnt mention any range , the output will be in between 0-1 [excluding 0 and 1]
print(random.random())
```

```python
#choice command

#gives a random number from the given array

x= random.choice([3,5,7,9])

print(x)
```

```python
x= random.choice([3,5,7,9], size=(2,4))

print(x)
```

```python
#permutations
#shuffle and permutation

a= np.array([2,4,6,8])
random.shuffle(a) #print(random.shuffle(a))--returns none
print(a)

a1 = np.arange(4).reshape(2,2)
np.random.shuffle(a1)
print("shuffle with shape", a1)


b= np.array([2,4,6,8])
c=np.random.permutation(b)
print(c)
print(b)


#arange(size)
arr = np.arange(12).reshape((3, 4))
np.random.permutation(arr)
```

```python
#multi dimentional arrays are shuffled only along first axis
arr = np.arange(9).reshape((3, 3))
np.random.shuffle(arr)
arr
```

```python
#Normal distribution
#(mean,sd,size)

x= np.random.normal(size=(2,3))
print(x)


y= np.random.normal(loc=2, scale = 1.5,size=(2,3))
print(y)
```

```python
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
#distplot - distribution plot

#Normal distribution: continuos

y= np.random.normal(loc=2, scale = 1.5,size=(2,3))
print(y)

sns.distplot(y, hist=False)

plt.show()
```

```python
sns.distplot([1,2,3,4], hist = True)

plt.show()
```

```python
#Binomial distribution (no of trials, probability, shape) : discrete

y= np.random.binomial(n=50,p=0.5,size=50)
print(y)

sns.distplot(y, hist=False)

```

```python
sns.kdeplot(np.random.normal(loc=50, scale=5, size=1000),label='normal')
sns.kdeplot(np.random.binomial(n=100, p=0.5, size=1000),label='binomial')

plt.show()
```

```python
#poisson distribution: discrete
#lam - known no of occurances

x= np.random.poisson(lam= 5, size = 10)
print(x)

sns.histplot(x)
```

```python
#uniform distribution : continuos
#(lowerbound = 0 , upperbound= 1, size)


x= np.random.uniform(0,1,size = 10)
print(x)

sns.distplot(x, hist=False)
```

```python
#Exponential function()

sns.distplot(np.random.exponential(scale=2, size=100), hist=False)


#Poisson distribution deals with number of occurences of an event in a time period 
#whereas exponential distribution deals with the time between these events.
```
